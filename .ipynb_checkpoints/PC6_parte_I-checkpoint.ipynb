```python
airports=gpd.read_file(linkCanada,layer='airports')
rivers=gpd.read_file(linkCanada,layer='rivers')
country=gpd.read_file(linkCanada,layer='country')
```

```python
import pandas as pd
portsFileLink="https://github.com/CRVCh-16/PC6/raw/refs/heads/main/UpdatedPub150.csv"
infoseaports=pd.read_csv(portsFileLink)
```

```python
# nos quedamos con las columnas m√°s importantes
infoseaports.rename(columns={'Main Port Name':'portName'},inplace=True)
infoseaports=infoseaports.loc[:,['portName', 'Country Code','Latitude', 'Longitude']]
infoseaports.info()
```

```python
infoseaports[infoseaports["Country Code"]=="Canada"]
```

```python
# creamos un gdf a partir del df
seaports=gpd.GeoDataFrame(data=infoseaports.copy(), geometry=gpd.points_from_xy(infoseaports.Longitude, infoseaports.Latitude),crs=4326)

# nos quedamos con los seaports de Canada
seaports_can=seaports[seaports['Country Code']=='Canada'].copy()

# reseteamos los indices
seaports_can.reset_index(drop=True, inplace=True)

# reproyectamos a ESRI: 102001
seaports_can_102001=seaports_can.to_crs("ESRI:102001")
```

```python
seaports_can_102001.head()
```

```python
# ahora filtramos solo los aeropuertos m√°s grandes en canada
largeAirports=airports[airports['kind']=='large_airport']
largeAirports.reset_index(drop=True, inplace=True)

# se grafican los large airports encima de los seaports de Canada
base=largeAirports.plot(color='red',marker="^")
seaports_can_102001.plot(ax=base,alpha=0.5,markersize=3)
```

```python
# ahora calculamos las distancias entre los seaports y los airports de Canada

seaports_can_102001.set_index('portName').geometry.apply\
(lambda g: largeAirports.set_index('name').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)
```

```python
# guardamos esta √∫ltima matriz
distanceMatrixKM_sea_air=seaports_can_102001.set_index('portName').geometry.apply\
(lambda g: largeAirports.set_index('name').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)
```

```python
# ahora escogemos la 1ra fila (seaport=Ahousat)
distanceMatrixKM_sea_air.iloc[0].sort_values(ascending=True)
```

## Ejercicio 3
1. Create a HULL for some set of line map.
2. Compute the distance matrix between the HULLS and a map of points.
3. Plot the HULLS and the points. Show the closest and farthest points to the HULL.

```python
# OJO solo mostrar los mas cercanos y mas lejanos, no todos
```

```python
rivers[~rivers.SYSTEM.isna()]
```

```python
systems=rivers.dissolve(by='SYSTEM')
systems
```

```python
# formateamos el GDF

systems.reset_index(drop=False,inplace=True)
systems.drop(columns='NAME',inplace=True)
systems
```

```python
systems.convex_hull.plot()
```

```python
systems_hulls=systems.convex_hull.to_frame()
systems_hulls['system']=['Mackenzie']
systems_hulls.rename(columns={0:'geometry'},inplace=True)
systems_hulls=systems_hulls.set_geometry('geometry')
systems_hulls.crs="ESRI:102001"
systems_hulls
```

```python
distanceMatrixKM_sysHull_air=systems_hulls.set_index('system').geometry.apply\
(lambda g: largeAirports.set_index('name').geometry.distance(g)/1000).\
sort_index(axis=0).sort_index(axis=1)

distanceMatrixKM_sysHull_air
```

```python
distanceMatrixKM_sysHull_air.loc["Mackenzie"].sort_values(ascending=True)
```

```python
base=country.plot(facecolor='#feffe5',edgecolor='black')
systems_hulls.plot(ax=base,facecolor='green',edgecolor='white',alpha=0.4,hatch='X')
largeAirports[largeAirports.name.str.contains("Edmonton|John's")].plot(ax=base,color='red',marker="^")
```

```python
base=largeAirports[largeAirports.name.str.contains("Edmonton|John's")].explore(color='red',marker_kwds=dict(radius=5))
systems_hulls.explore(m=base)
```

## Ejercicio 4
1. Select a line map and a point one.
2. Get the buffer for the lines, select a distance.
3. Keep the points that are within the buffer (you might need to play with differn distances until you show something interesting).

```python
distanceMatrixKM_riv_air
```

```python
small_airports=airports[airports['kind']=='small_airport']
```

```python
# para la distncia para el  buffer, seleecionamos el minimo del minimo
distanceMatrixKM_riv_air.min(axis=1).min()
```

```python
# el buffer ser√° 5 veces la distancia minima, como en el ejemplo mostrado en clase
minMinMts_5=5*distanceMatrixKM_riv_air.min(axis=1).min()*1000
# se crea el buffer para todos los rios
riversAll_buf=rivers.buffer(distance=minMinMts_5)
riversAll_buf
```

```python
# nos quedamos solo con los small airports dentro de los buffer
allSmallAirportsWithinBuffs=small_airports.clip(riversAll_buf)
allSmallAirportsWithinBuffs
```

```python
# ploteamos
base=country.plot(facecolor='#feffe5',edgecolor='black')
riversAll_buf.plot(ax=base,color='yellow')
allSmallAirportsWithinBuffs.plot(ax=base, color='green', markersize=1)
rivers.plot(ax=base,color='blue',linewidth=0.5) # no deja ver bien los airports y los buffer
```

```python
base=riversAll_buf.explore(color='yellow')
allSmallAirportsWithinBuffs.explore(m=base,color='green')
rivers.explore(m=base,color='blue',style_kwds={'weight':0.5})
```
